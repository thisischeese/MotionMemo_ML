```python
import os

# 루트 폴더 경로 설정
root_folder_path = 'your root path'

# 폴더 내의 4단계 하위 폴더 수를 저장할 딕셔너리
folder_lvl4_count = {}

# 루트 폴더 내의 각 폴더에 대해 반복
for folder_name_lvl1 in os.listdir(root_folder_path):
    folder_path_lvl1 = os.path.join(root_folder_path, folder_name_lvl1)
    if os.path.isdir(folder_path_lvl1):
        # 1단계 하위 폴더에 대해 반복
        for folder_name_lvl2 in os.listdir(folder_path_lvl1):
            folder_path_lvl2 = os.path.join(folder_path_lvl1, folder_name_lvl2)
            if os.path.isdir(folder_path_lvl2):
                # 2단계 하위 폴더에 대해 반복
                for folder_name_lvl3 in os.listdir(folder_path_lvl2):
                    folder_path_lvl3 = os.path.join(folder_path_lvl2, folder_name_lvl3)
                    if os.path.isdir(folder_path_lvl3):
                        # 3단계 하위 폴더에 대해 반복
                        lvl4_count = 0
                        for folder_name_lvl4 in os.listdir(folder_path_lvl3):
                            folder_path_lvl4 = os.path.join(folder_path_lvl3, folder_name_lvl4)
                            if os.path.isdir(folder_path_lvl4):
                                lvl4_count += 1
                        folder_lvl4_count[folder_path_lvl3] = lvl4_count

# 결과 출력
for folder_path_lvl3, lvl4_count in folder_lvl4_count.items():
  if(lvl4_count!=10):
    print(f"Error : '{folder_path_lvl3}' 내의 4단계 하위 폴더 수: {lvl4_count}개")

```

```python
import os


error=0
# 루트 폴더 내의 각 폴더에 대해 반복
for folder_name_lvl1 in os.listdir(root_folder_path):
    folder_path_lvl1 = os.path.join(root_folder_path, folder_name_lvl1)
    if os.path.isdir(folder_path_lvl1):
        wp_folder_cnt=0
        # 1단계 하위 폴더에 대해 반복
        for folder_name_lvl2 in os.listdir(folder_path_lvl1):
            folder_path_lvl2 = os.path.join(folder_path_lvl1, folder_name_lvl2)
            if os.path.isdir(folder_path_lvl2):
                wp_folder_cnt+=1
                label_folder_cnt=0
                # 2단계 하위 폴더에 대해 반복
                for folder_name_lvl3 in os.listdir(folder_path_lvl2):
                    folder_path_lvl3 = os.path.join(folder_path_lvl2, folder_name_lvl3)
                    if os.path.isdir(folder_path_lvl3):
                      label_folder_cnt+=1
                # 3단계 하위 폴더(10가지 라벨)가 10가지 모두 다 있는 경우
                if(label_folder_cnt!=10):
                  print(f"Error : In '{folder_name_lvl2}' some label folders are lost.")
                  error+=1
        if(wp_folder_cnt!=2):
          print(f"Error : In '{folder_name_lvl1}' 'words' folder or 'phrases' folder is lost")
          error+=1
if (error!=0):
  print(f"'{error}' error occured")
else:
  print("No error occured")
```

```python
# 폴더 수를 저장할 변수 초기화
folder_count = 0

# 루트 폴더 내의 각 항목에 대해 반복
for item in os.listdir(root_folder_path):
    item_path = os.path.join(root_folder_path, item)
    if os.path.isdir(item_path):
        folder_count += 1

# 결과 출력
print(f"폴더 '{root_folder_path}' 아래에 총 {folder_count}개의 폴더가 있습니다.")
```

```python
import os


# 가장 큰 4단계 하위 폴더 개수 저장 변수 초기화
max_folder_count = 0

# 루트 폴더 내의 각 폴더에 대해 반복
for folder_name_lvl1 in os.listdir(root_folder_path):
    folder_path_lvl1 = os.path.join(root_folder_path, folder_name_lvl1)
    if os.path.isdir(folder_path_lvl1):
        # 1단계 하위 폴더에 대해 반복
        for folder_name_lvl2 in os.listdir(folder_path_lvl1):
            folder_path_lvl2 = os.path.join(folder_path_lvl1, folder_name_lvl2)
            if os.path.isdir(folder_path_lvl2):
                # 2단계 하위 폴더에 대해 반복
                for folder_name_lvl3 in os.listdir(folder_path_lvl2):
                    folder_path_lvl3 = os.path.join(folder_path_lvl2, folder_name_lvl3)
                    if os.path.isdir(folder_path_lvl3):
                        # 3단계 하위 폴더에 대해 반복
                        max_lvl4_count = 0
                        for folder_name_lvl4 in os.listdir(folder_path_lvl3):
                            folder_path_lvl4 = os.path.join(folder_path_lvl3, folder_name_lvl4)
                            if os.path.isdir(folder_path_lvl4):
                                # 4단계 하위 폴더 개수 세기
                                lvl4_count = len(os.listdir(folder_path_lvl4))
                                if lvl4_count > max_lvl4_count:
                                    max_lvl4_count = lvl4_count
                        # 현재 폴더의 최대 4단계 하위 폴더 개수가 가장 큰 값인지 확인
                        if max_lvl4_count > max_folder_count:
                            max_folder_count = max_lvl4_count

# 최대 4단계 하위 폴더 개수 출력
print("가장 큰 4단계 하위 폴더 개수:", max_folder_count)
```
